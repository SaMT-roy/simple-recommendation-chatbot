```python
import string
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
from newspaper import Article
```

```python
article=Article('https://www.mayoclinic.org/diseases-conditions/heart-disease/symptoms-causes/syc-20353118')
```

```python
article.download()
article.parse()
article.nlp()

corpus = article.text
```

```python
print(corpus)
```

```python
text = corpus
sent = nltk.sent_tokenize(text)
```

```python
print(sent)
```

```python
def resp_greet(text):
    text = text.lower()
    
    greetings_msg_bot = ['hey','hello']
    greetings_msg_user = ['hi','hello']
    
    for word in text.split():
        if word in greetings_msg_user:
            
            return np.random.choice(greetings_msg_bot)
```

```python
def sort_index(list_var):
    
    length = len(list_var)
    list_index = list(range(0,length))
    
    y = list_var
    
    for i in range(length):
        for j in range(length):
            if y[list_index[i]] > y[list_index[j]]:
                temp = list_index[i]
                list_index[i] = list_index[j]
                list_index[j] = temp
    return list_index
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
    
for i in range(0, len(sent)):
        review = sent[i].lower()
        review = review.split()
    
        review = [lemmatizer.lemmatize(word) for word in review ]
        review = ' '.join(review)
        corpus.append(review)
```

```python
string.punctuation
```

```python
x = [''.join(c for c in s if c not in string.punctuation) for s in corpus]
```

```python
x
```

```python
def response_bot(user_input):
    user_input = user_input.lower()
    x.append(user_input)
    response_bot = ''
    
    cv = CountVectorizer().fit_transform(x)
    
    similarity_score = cosine_similarity(cv[-1],cv)
    similarity_score_list = similarity_score.flatten()
    
    index = sort_index(similarity_score_list)
    index = index[1:]
    response_flag = 0
    
    j = 0
    for i in range(len(index)):
        if similarity_score_list[index[i]] > 0.0:
            response_bot = response_bot+''+ sent[index[i]]
            response_flag = 1
            j=j+1
        if j>20:
            break
    
    if response_flag == 0:
        response_bot = response_bot+''+'i dont`t understand'
    x.remove(user_input)
    return response_bot
```

```python
print('Bot: I am a bot')
exit = ['exit','bye']

while(True):
    print('Enter exit or bye or hi or hello : ')
    user_input = input()
    if user_input.lower() in exit:
        print('bot: chat with you later')
        break
    
    if resp_greet(user_input) != None:
        print('bot :'+resp_greet(user_input))
    else:
        print('bot :'+response_bot(user_input))
```
